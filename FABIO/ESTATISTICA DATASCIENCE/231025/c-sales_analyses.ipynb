```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Coffe_sales.csv')
sales = df['money']

df['Date'] = pd.to_datetime(df['Date'])
df['Date'] = df['Date'].dt.strftime('%d/%m/%Y')
df
```

```python
mobile_mean_sales = pd.Series(sales).rolling(window=30).mean()

mobile_mean_sales
```

```python
fig, ax = plt.subplots()
ax.plot(mobile_mean_sales, marker='o', color='blue', label='Media movel' )
ax.plot(sales, marker='o', color='red', label='Media movel' )
ax.set_xlabel("Dia da semana")
ax.set_ylabel("Vendas (unidade)")
ax.set_title("Vendas diarias")
ax.set_ylim(min(sales), max(sales))
ax.set_xticks(range(len(df['Date'])))
ax.set_xticklabels(df['Date'], rotation=45)
ax.legend(loc='upper left', )
plt.show()
```
